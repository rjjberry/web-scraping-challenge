```python
# Dependencies
import pymongo
import datetime
```

```python
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Declare the database
db = client.mars_db

# Declare the collection
collection = db.mars_db
```

```python
# Part I
# A dictionary that represents the document to be inserted
post = {
    'temperature (high)': '33.4 degrees F',
    'temperature (avg)': '-73.0 degrees F',
    'temperature (low)': '-135.5 degrees F',
    'wind speed (max)': '18.9 m/s',
    'wind speed (avg)': '5.0 m/s',
    'wind speed (min)': '0.3 m/s',
    'pressure (max)': 725.5,
    'pressure (avg)': 708.6,
    'pressure (min)': 684.8,
    'date': datetime.datetime.utcnow()
}
# Insert the document into the database
# The database and collection, if they don't already exist, will be created at this point.
#collection.insert_one(post)
```

```python
# Insert the document into the database
# Insert document into collection
collection.insert_one(post)
```

```python
# Verify results:
results = collection.find()
for result in results:
    print(result)
```
